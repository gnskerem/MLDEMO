```python
# =====================================================================
# ADIM 1: VERİ OKUMA VE SIMPLEIMPUTER İLE EKSİK VERİ YÖNETİMİ (Lab-09)
# =====================================================================
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer  # Lab-09 kütüphanesi

# Veri setini okuma
ev_veri = pd.read_csv('EV_Usage_Dataset_v2.csv')

print("--- Adım 1: Veri Kümesine Genel Bakış ---")
print(f"Orijinal Satır Sayısı: {ev_veri.shape[0]}, Sütun Sayısı: {ev_veri.shape[1]}")

# Feature Reduction: Anlamsız kimlik ve tarih verilerinin düşürülmesi (Lab-03)
gereksiz_sutunlar = ['user_id', 'trip_date']
ev_veri = ev_veri.drop(columns=[col for col in gereksiz_sutunlar if col in ev_veri.columns], axis=1)

# Veri tiplerine göre kolonların ayrılması
sayisal_kolonlar = ev_veri.select_dtypes(include=[np.number]).columns.tolist()
kategorik_kolonlar = ev_veri.select_dtypes(include=['object']).columns.tolist()

# Lab-09: Sayısal veriler için medyan imputer uygulaması
sayisal_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
ev_veri[sayisal_kolonlar] = sayisal_imputer.fit_transform(ev_veri[sayisal_kolonlar])

# Lab-09: Kategorik veriler için en sık tekrar eden (most frequent) imputer uygulaması
kategorik_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
ev_veri[kategorik_kolonlar] = kategorik_imputer.fit_transform(ev_veri[kategorik_kolonlar])

print(f"İşlem Sonrası Kalan Toplam Eksik Veri Sayısı: {ev_veri.isnull().sum().sum()}")
```

```python
# ADIM 2: MANUEL ENCODING

gelir_sozlugu = {'low': 0, 'middle': 1, 'high': 2}
if 'user_income_level' in ev_veri.columns:
    ev_veri['user_income_level'] = ev_veri['user_income_level'].map(gelir_sozlugu)

# İki sınıflı metin kolonlarının binary kodlara dönüştürülmesi
for kolon in ev_veri.select_dtypes(include=['object']).columns:
    if ev_veri[kolon].nunique() <= 2:
        ev_veri[kolon] = ev_veri[kolon].astype('category').cat.codes

# Çok sınıflı kategorik değişkenler için One-Hot Encoding uygulanması
ev_veri = pd.get_dummies(ev_veri, drop_first=True)
ev_veri = ev_veri.astype(float)

print("--- Adım 2: Kategorik Verilerin Dönüşümü Tamamlandı ---")
print(f"Güncel Sütun Sayısı: {ev_veri.shape[1]}")
```

```python
# =====================================================================
# ADIM 3: VERİ DAĞILIMLARI VE ISI HARİTASI ANALİZİ (Lab-01 & Ödev Şartı)
# =====================================================================
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("ticks")
```

```python
# Şekil 3: Araç batarya kapasitelerinin genel frekans dağılımı
plt.figure(figsize=(7, 4))
sns.histplot(ev_veri['battery_capacity_kwh'], bins=30, kde=True, color='dodgerblue')
plt.title('Sürücü Araçlarının Batarya Kapasitesi Dağılımı')
plt.xlabel('Batarya Kapasitesi (kWh)')
plt.ylabel('Frekans')
plt.show()
```

```python
# =====================================================================
# ADIM 4: VERİ NORMALİZASYONU VE LAB-02 TARZI PROJE AYRIMI
# =====================================================================
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Lab-02 Slayt 24: .drop ve .loc ile hedef ve nitelik ayırma mantığı
X = ev_veri.drop('range_anxiety_risk', axis=1)
Y = ev_veri.loc[:, 'range_anxiety_risk']

# Ödev Şartı (Figure 5): MinMaxScaler ile 0-1 Normalizasyonu
scaler = MinMaxScaler()
X_normalize = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Verinin Eğitim ve Test kümelerine bölünmesi
egitimX, testX, egitimY, testY = train_test_split(X_normalize, Y, test_size=0.2, random_state=42)

print("--- Adım 4: Veri Bölme ve Ölçekleme Başarılı ---")
print(f"Eğitim Kümesi Satır Sayısı: {egitimX.shape[0]}, Test Kümesi Satır Sayısı: {testX.shape[0]}")
```

```python
# =====================================================================
# ADIM 5: 5 FARKLI MAKİNE ÖĞRENMESİ MODELİNİN EĞİTİLMESİ (Lab-04)
# =====================================================================
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Lab-04 Slayt 34: random_state=13 parametre kullanım alışkanlığı
ml_modelleri = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=13),
    "K-Nearest Neighbors (KNN)": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree Classifier": DecisionTreeClassifier(random_state=13),
    "Random Forest Classifier": RandomForestClassifier(random_state=13),
    "AdaBoost Classifier": AdaBoostClassifier(random_state=13)
}

print("--- Adım 5: Modellerin Orijinal Veriyle Eğitimi ---")
en_iyi_ml_skoru = 0
en_iyi_ml_adi = ""

for isim, model in ml_modelleri.items():
    model.fit(egitimX, egitimY)
    tahminler = model.predict(testX)
    skor = accuracy_score(testY, tahminler)
    print(f"> {isim} Doğruluk Oranı (Accuracy): {skor:.4f}")

    if skor > en_iyi_ml_skoru:
        en_iyi_ml_skoru = skor
        en_iyi_ml_adi = isim

print(f"\n[EN BAŞARILI MODEL]: {en_iyi_ml_adi} (Doğruluk: {en_iyi_ml_skoru:.4f})")
```

```python
# =====================================================================
# ADIM 6: XGBOOST ENTEGRASYONU, SMOTE VE HATA MATRİSLERİ (Lab-05 & Lab-06)
# =====================================================================
from xgboost import XGBClassifier  # Lab-05 Slayt 56 Özel Vurgusu
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, classification_report

# Listeye XGBoost modelini de ekleyip eğitiyoruz
xgboost_model = XGBClassifier(random_state=13, eval_metric='logloss')
xgboost_model.fit(egitimX, egitimY)
xgb_skor = accuracy_score(testY, xgboost_model.predict(testX))
print(f"> Gelişmiş XGBoost Classifier Doğruluk Oranı: {xgb_skor:.4f}")

# Eğer XGBoost en iyisiyse dinamik olarak atıyoruz
if xgb_skor > en_iyi_ml_skoru:
    ml_modelleri["XGBoost Classifier"] = xgboost_model
    en_iyi_ml_adi = "XGBoost Classifier"

en_iyi_model_nesnesi = ml_modelleri[en_iyi_ml_adi]

# SMOTE Öncesi Hata Matrisi Görselleştirmesi
smote_oncesi_tahmin = en_iyi_model_nesnesi.predict(testX)
cm_oncesi = confusion_matrix(testY, smote_oncesi_tahmin)

plt.figure(figsize=(4.5, 3.5))
sns.heatmap(cm_oncesi, annot=True, fmt="d", cmap="Oranges", cbar=False)
plt.title(f"SMOTE Öncesi - {en_iyi_ml_adi}\nHata Matrisi (Confusion Matrix)")
plt.ylabel('Gerçek Sınıf')
plt.xlabel('Tahmin Edilen Sınıf')
plt.show()

# Ödev Şartı: Yalnızca eğitim kümesine SMOTE uygulanarak veri dengelenmesi
smote = SMOTE(random_state=13)
egitimX_smote, egitimY_smote = smote.fit_resample(egitimX, egitimY)

# En başarılı modelin dengelenmiş veriyle yeniden eğitilmesi
en_iyi_model_nesnesi.fit(egitimX_smote, egitimY_smote)
smote_sonrasi_tahmin = en_iyi_model_nesnesi.predict(testX)
cm_sonrasi = confusion_matrix(testY, smote_sonrasi_tahmin)

# SMOTE Sonrası Hata Matrisi Görselleştirmesi
plt.figure(figsize=(4.5, 3.5))
sns.heatmap(cm_sonrasi, annot=True, fmt="d", cmap="GnBu", cbar=False)
plt.title(f"SMOTE Sonrası - {en_iyi_ml_adi}\nHata Matrisi (Confusion Matrix)")
plt.ylabel('Gerçek Sınıf')
plt.xlabel('Tahmin Edilen Sınıf')
plt.show()

# Lab-05 Slayt 12: Akademik Sınıflandırma Raporu (Precision, Recall, F1)
print(f"\n--- {en_iyi_ml_adi} SMOTE Sonrası Detaylı Başarım Raporu ---")
print(classification_report(testY, smote_sonrasi_tahmin))
```

```python
# =====================================================================
# ADIM 7: 3 FARKLI DERİN ÖĞRENME MİMARİSİNİN EĞİTİLMESİ (Lab-07 & Lab-10)
# =====================================================================
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation

giris_boyutu = egitimX_smote.shape[1]

# MODEL 1: Standart Yapay Sinir Ağı (Lab-07 Giriş Seviyesi)
model1 = Sequential()
model1.add(Dense(32, input_shape=(giris_boyutu,)))
model1.add(Activation('relu'))
model1.add(Dense(1))
model1.add(Activation('sigmoid'))
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# MODEL 2: Gelişmiş Derin Yapay Sinir Ağı (DNN - Lab-10 Çok Katmanlı Yapı)
model2 = Sequential()
model2.add(Dense(64, input_shape=(giris_boyutu,)))
model2.add(Activation('relu'))
model2.add(Dense(32))
model2.add(Activation('relu'))
model2.add(Dense(16))
model2.add(Activation('relu'))
model2.add(Dense(1))
model2.add(Activation('sigmoid'))
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# MODEL 3: İteratif İyileştirilmiş Düzenli DNN (Dropout Entegrasyonu - Lab-08/10)
model3 = Sequential()
model3.add(Dense(64, input_shape=(giris_boyutu,)))
model3.add(Activation('relu'))
model3.add(Dropout(0.3))  # Ezberlemeyi önleme katmanı

model3.add(Dense(32))
model3.add(Activation('relu'))
model3.add(Dropout(0.3))

model3.add(Dense(16))
model3.add(Activation('relu'))

model3.add(Dense(1))
model3.add(Activation('sigmoid'))
model3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("--- Adım 7: Derin Öğrenme Modelleri Çalıştırılıyor (Batch: 128) ---")
gecmis1 = model1.fit(egitimX_smote, egitimY_smote, batch_size=128, epochs=10, validation_data=(testX, testY), verbose=0)
gecmis2 = model2.fit(egitimX_smote, egitimY_smote, batch_size=128, epochs=12, validation_data=(testX, testY), verbose=0)
# Lab-10 Slayt 45: 15 Dönemlik (Epoch) ileri seviye eğitim takibi
gecmis3 = model3.fit(egitimX_smote, egitimY_smote, batch_size=128, epochs=15, validation_data=(testX, testY), verbose=1)
```

```python
# =====================================================================
# ADIM 8: MODEL KIYASLAMA TABLOSU (Table 7) VE ROC EĞRİLERİ (Figure 11)
# =====================================================================
import time
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc

sonuclar = {}

# Klasik Makine Öğrenmesi Modellerinin Zaman ve Performans Çıkarımı
for isim, model in ml_modelleri.items():
    baslangic = time.time()
    tahminler = model.predict(testX)
    bitis = time.time()

    # Tek bir kayıt için geçen sürenin hesabı (Runtime per record)
    tek_kayit_sure = (bitis - baslangic) / len(testX)
    olasiliklar = model.predict_proba(testX)[:, 1] if hasattr(model, "predict_proba") else tahminler

    fpr, tpr, _ = roc_curve(testY, olasiliklar)
    sonuclar[isim] = {
        "Accuracy": accuracy_score(testY, tahminler),
        "Precision": precision_score(testY, tahminler, zero_division=0),
        "Recall": recall_score(testY, tahminler, zero_division=0),
        "F1-Score": f1_score(testY, tahminler, zero_division=0),
        "AUC": auc(fpr, tpr),
        "Runtime_Per_Rec": tek_kayit_sure,
        "FPR": fpr, "TPR": tpr
    }

# Geliştirilmiş Derin Öğrenme Modelinin (Model 3) Analizi
baslangic = time.time()
dl_olasilik = model3.predict(testX, verbose=0).flatten()
bitis = time.time()

dl_sinif = (dl_olasilik > 0.5).astype(int)
dl_fpr, dl_tpr, _ = roc_curve(testY, dl_olasilik)

sonuclar["Deep Learning (Model 3)"] = {
    "Accuracy": accuracy_score(testY, dl_sinif),
    "Precision": precision_score(testY, dl_sinif, zero_division=0),
    "Recall": recall_score(testY, dl_sinif, zero_division=0),
    "F1-Score": f1_score(testY, dl_sinif, zero_division=0),
    "AUC": auc(dl_fpr, dl_tpr),
    "Runtime_Per_Rec": (bitis - baslangic) / len(testX),
    "FPR": dl_fpr, "TPR": dl_tpr
}

# Akademik Karşılaştırma Tablosunun (Table 7) Gösterimi
df_performans = pd.DataFrame(sonuclar).T.drop(columns=['FPR', 'TPR'])
print("\n=== TABLO 7: TÜM MODELLERİN AKADEMİK KIYASLAMASI ===")
print(df_performans)

# Şekil 11: ROC Eğrilerinin Çizilmesi
plt.figure(figsize=(7.5, 5.5))
for isim, metrikler in sonuclar.items():
    plt.plot(sonuclar[isim]["FPR"], sonuclar[isim]["TPR"], label=f'{isim} (AUC = {sonuclar[isim]["AUC"]:.2f})')
plt.plot([0, 1], [0, 1], 'k--', color='gray')
plt.xlabel('False Positive Rate (YVO)')
plt.ylabel('True Positive Rate (DOO)')
plt.title('Şekil 11: Modellerin ROC Eğrileri Karşılaştırması')
plt.legend(loc="lower right")
plt.grid(True, linestyle=':')
plt.show()
```

```python
# =====================================================================
# ADIM 9: MODEL VE SCALER NESNELERİNİN DİSKE KAYDEDİLMESİ (Lab-08)
# =====================================================================
import joblib

# Eğitilmiş nesnelerin dosya sistemine aktarılması
joblib.dump(ml_modelleri[en_iyi_ml_adi], 'en_iyi_ev_ml_model.pkl')
joblib.dump(scaler, 'ev_minmax_scaler.pkl')

print(f"[BAŞARILI] {en_iyi_ml_adi} nesnesi ve normalleştirici lokal sisteme kaydedildi.")
```

```python
# =====================================================================
# ADIM 10: GRADIO KÜTÜPHANESİ İLE CANLI DEMO ARAYÜZÜ (GUI Development)
# =====================================================================
!pip install -q gradio

import gradio as gr
import joblib

# Kaydedilen nesnelerin geri yüklenmesi (Load Saved Models Kuralı)
model_gui = joblib.load('en_iyi_ev_ml_model.pkl')
scaler_gui = joblib.load('ev_minmax_scaler.pkl')
ham_veri = pd.read_csv('EV_Usage_Dataset_v2.csv')

def anlik_menzil_kaygisi_tahmin(sehir, batarya_kapasite, batarya_saglik, mesafe, gunluk_sefer, sarj_frekans, sarj_turu, sarj_sure, istasyon_mesafe, elektrik_maliyet, hava_durumu, trafik_yogunluk, gelir_seviyesi, tahmini_menzil):
    # Modelin eğitildiği kolon formatında boş bir satır oluşturma
    girdi_satiri = pd.DataFrame(0, index=[0], columns=X.columns)

    # Sayısal alanların doldurulması
    girdi_satiri['battery_capacity_kwh'] = float(batarya_kapasite)
    girdi_satiri['battery_health_pct'] = float(batarya_saglik)
    girdi_satiri['distance_km'] = float(mesafe)
    girdi_satiri['daily_trip_count'] = float(gunluk_sefer)
    girdi_satiri['charging_frequency_per_week'] = float(sarj_frekans)
    girdi_satiri['charging_duration_min'] = float(sarj_sure)
    girdi_satiri['charging_station_distance_km'] = float(istasyon_mesafe)
    girdi_satiri['electricity_cost_per_kwh'] = float(elektrik_maliyet)
    girdi_satiri['traffic_density'] = float(trafik_yogunluk)
    girdi_satiri['range_km_estimated'] = float(tahmini_menzil)
    if 'year' in girdi_satiri.columns: girdi_satiri['year'] = 2026.0
    if 'month' in girdi_satiri.columns: girdi_satiri['month'] = 5.0

    # Lab-03 Tarzı manuel gelir eşlemesi
    gelir_kodlari = {'low': 0, 'middle': 1, 'high': 2}
    girdi_satiri['user_income_level'] = gelir_kodlari[gelir_seviyesi]

    # Kategorik alanların One-Hot durumuna göre işaretlenmesi
    for sütun in girdi_satiri.columns:
        if f"city_{sehir}" in sütun: girdi_satiri[f"city_{sehir}"] = 1.0
        if f"charging_type_{sarj_turu}" in sütun: girdi_satiri[f"charging_type_{sarj_turu}"] = 1.0
        if f"weather_condition_{hava_durumu}" in sütun: girdi_satiri[f"weather_condition_{hava_durumu}"] = 1.0

    # Normalizasyon ve Tahmin adımları
    girdi_norm = scaler_gui.transform(girdi_satiri)
    tahmin_sinif = model_gui.predict(girdi_norm)[0]
    ihtimaller = model_gui.predict_proba(girdi_norm)[0]
    risk_yuzdesi = ihtimaller[1] * 100

    if tahmin_sinif == 1:
        return f"🔴 TAHMİN: YÜKSEK MENZİL KAYGISI RİSKİ!\nSürücünün yolda kalma korkusu yaşama ihtimali: %{risk_yuzdesi:.2f}"
    else:
        return f"🟢 TAHMİN: DÜŞÜK RİSK / GÜVENLİ SÜRÜŞ\nSürücünün endişe yaşama ihtimali: %{risk_yuzdesi:.2f}"

# Arayüz elemanlarının tanımlanması
arayuz = gr.Interface(
    fn=anlik_menzil_kaygisi_tahmin,
    inputs=[
        gr.Dropdown(list(ham_veri['city'].dropna().unique()), label="Şehir"),
        gr.Number(value=35.0, label="Batarya Kapasitesi (kWh)"),
        gr.Slider(50, 100, value=85, label="Batarya Sağlık Oranı (%)"),
        gr.Number(value=50.0, label="Sürüş Mesafesi (km)"),
        gr.Number(value=2.0, label="Günlük Sefer Sayısı"),
        gr.Slider(1, 14, value=4, step=1, label="Haftalık Şarj Sıklığı"),
        gr.Dropdown(list(ham_veri['charging_type'].dropna().unique()), label="Şarj Tipi"),
        gr.Number(value=80, label="Şarj Süresi (Dakika)"),
        gr.Number(value=3.0, label="İstasyona Olan Mesafe (km)"),
        gr.Number(value=6.8, label="Elektrik Birim Maliyeti"),
        gr.Dropdown(list(ham_veri['weather_condition'].dropna().unique()), label="Hava Koşulu"),
        gr.Slider(0.0, 1.0, value=0.4, label="Trafik Yoğunluğu"),
        gr.Dropdown(['low', 'middle', 'high'], label="Sürücü Gelir Grubu"),
        gr.Number(value=110.0, label="Gösterge Panelindeki Tahmini Menzil (km)")
    ],
    outputs=gr.Textbox(label="Yapay Zeka Karar Raporu", lines=3),
    title="Elektrikli Araç Kullanıcıları İçin Yapay Zeka Tabanlı Menzil Kaygısı Analiz Sistemi",
    description="Sürücü ve araç verilerini girerek modelimizin anlık kararlarını gözlemleyebilirsiniz. Hocanıza demo yaparken share=True linkini kullanabilirsiniz.",
    theme="soft"
)

# Canlı sunum linki üretimi için share=True tetikleniyor
arayuz.launch(share=True)
```
